```python
#import numpy library to calculate complex polynomial equations with no straightforward defined formula (cubic & quartic)
import numpy

#Define function for quadractic equation formula and collect inputs A, B, and C to calculate the roots
def quadractic_equation() :
    A = int(input('What is the coefficient of x^2 - Input whole numbers only: '))
    B = int(input('What is the co-efficient of x - Input whole numbers only: '))
    C = int(input('What is  the constant (C): '))
    
    #Define expressions to calculate the roots
    roots_1 = (- B + ((B**2 - 4*A*C)**0.5)) / 2*A
    roots_2 = (- B - ((B**2 - 4*A*C)**0.5)) / 2*A

    #Print answers of quadractic equation
    print (f"The roots of the quadratic equation are {roots_1} and {roots_2}")


#Define function for cubic equation formula and collect inputs A, B, C and D to calculate the roots
def cubic_equation() :
    A = int(input('What is the coefficient of x^3 - Input whole numbers only: '))
    B = int(input('What is the coefficient of x^2 - Input whole numbers only: '))
    C = int(input('What is the co-efficient of x - Input whole numbers only: '))
    D = int(input('What is  the constant (D): '))

    #Due to cubic equation calculation complexity, use numpy library to calculate cubic roots from user's input
    coefficients = [A,B,C,D]
    roots = numpy.roots(coefficients)

    #Print answers of cubic equation
    print (f"""
           The roots of the cubic equation are 
           {roots[-3]}, 
           {roots[-2]} and 
           {roots[-1]}""")


#Define function for quartic equation formula and collect inputs A, B, C, D and E to calculate the roots
def quartic_equation() :
    A = int(input('What is the coefficient of x^4 - Input whole numbers only: '))
    B = int(input('What is the coefficient of x^3 - Input whole numbers only: '))
    C = int(input('What is the coefficient of x^2 - Input whole numbers only: '))
    D = int(input('What is the co-efficient of x - Input whole numbers only: '))
    E = int(input('What is  the constant (E): '))
    

    #Due to quartic equation calculation complexity, use numpy library to calculate quartic roots from user's input
    coefficients = [A,B,C,D,E]
    roots = numpy.roots(coefficients)

    #Print answers of quartic equation
    print (f"""
           The roots of the quartic equation are 
           {roots[-4]}, 
           {roots[-3]} and 
           {roots[-2]}
           {roots[-1]}""")

#After defining functions for calculating all root equations that'll be present in the calculator, finally print user input form that collects what they'll like to calculate   
calculation = int(input('''
    Welcome to Prosperity's root calculator.
    What would you like to calculate today
    For Quadractic Equation, write 1
    For Cubic Equation, write 2
    For Quartic Equation, write 3: 
    '''
    ))

#Define logic for calling which the roots functions based on user input
if calculation == 1 :
    quadractic_equation()
    print("Thanks for using Prosperity's roots calculator, enjoy using your roots")
elif calculation == 2 :
    cubic_equation()
    print("Thanks for using Prosperity's roots calculator, enjoy using your roots")
elif calculation == 3 :
    quartic_equation()
    print("Thanks for using Prosperity's roots calculator, enjoy using your roots")
else :
    print ("Thanks for trying but it seems you used a wrong input or something went wrong, type CTRL + C and try again")
```
