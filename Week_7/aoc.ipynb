```python
with open("input.txt") as f:
    data = f.read().splitlines()
    
    sum = 0

    for i in data:
        parsed = i.split()

        for j in parsed:
            sort = sorted(j)
            counter = 0
            for k in range(len(sort)):
                print(k)
                if counter % 2 == 0:
                    for l in range(k):
                        print(i, j, k, l, sort[k], sort[k][l], sep=", ")
                        sum += sort[k+1][l] - sort[k][l]
                else:
                    continue
                counter += 1

    print(sum)
```
